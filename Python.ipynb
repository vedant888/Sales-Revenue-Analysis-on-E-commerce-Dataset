### **III.Type Casting**

```python
# Type Casting
df['Age'] = df['Age'].fillna(0).astype(int)
```

```python
df.duplicated()
```

### **V. Missing Values**

```python
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]  # Only show columns with missing values
missing_values
```

```python
# Rating Imputation
df['Rating'].fillna(df['Rating'].median(), inplace = True)
```

```python
df['Rating'].isnull().sum()
```

```python
# Shipment Time(days) Imputation
df['ShipmentTime(days)'].fillna(df['ShipmentTime(days)'].mean(), inplace=True)
```

```python
df['ShipmentTime(days)'].isnull().sum()
```

```python
df['ReturnReason'].isnull().sum()
```

```python
# Return Reason Imputation
df['ReturnReason'].fillna(df['ReturnReason'].mode()[0], inplace= True)
```

```python
df['ReturnReason'].isnull().sum()
```

### **VI. Outlier Treatment using BOX plott**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the plot style
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Age'])

# Add title and show the plot
plt.title("Box Plot of Age")
plt.xlabel("Age")
plt.grid(True)
plt.show()
```

```python
# Detect outliers in Age using IQR
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_age = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

# Show number and sample of outliers
print(f"Number of outliers in Age: {len(outliers_age)}")
print(outliers_age[['Age']].head())
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# Example with 'Age' column
plt.figure(figsize=(6, 6))
stats.probplot(df['Age'], dist="norm", plot=plt)
plt.title("Qâ€“Q Plot of Age")
plt.grid(True)
plt.show()
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# List of numeric columns to plot
columns_to_plot = ['Age', 'Revenue', 'ChurnRiskScore']

# Set up plot grid
plt.figure(figsize=(15, 4))

# Generate Q-Q plots
for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(1, 3, i)
    stats.probplot(df[column], dist="norm", plot=plt)
    plt.title(f"Qâ€“Q Plot of {column}")

plt.tight_layout()
plt.show()
```
